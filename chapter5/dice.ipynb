```python
import numpy as np
```

```python
def sample(dices=2):
    x = 0
    for _ in range(dices):
        x += np.random.choice([1, 2, 3, 4, 5, 6])
    return x
```

```python
print(sample())
print(sample())
print(sample())
```

```python
trial = 1000
V, n = 0, 0
```

```python
samples = []
for _ in range(trial) :
    s = sample()
    samples.append(s)
```

```python
V = sum(samples) / len(samples)
```

```python
print(V)
```

```python
trial = 1000
V, n = 0, 0
```

```python
for _ in range(trial):
    s = sample()
    n += 1
    V += (s - V) / n
    print(V)
```
